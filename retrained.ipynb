```python
from tensorflow.keras.models import load_model

# Load your saved ANN model
model = load_model(r'C:\Users\ABHINAV\Desktop\Heart_PA\models\ann_heart_disease_model.h5')

# Check exact input shape
print("Input shape expected:", model.input_shape)
print("Number of features  :", model.input_shape[1])
```

```python
import pandas as pd
import pickle

# Load your original heart.csv
df = pd.read_csv(r'C:\Users\ABHINAV\Desktop\Heart_PA\heart.csv')

# Apply same preprocessing as ANN notebook
df_encoded = pd.get_dummies(df, columns=[
    'Sex', 'ChestPainType', 'RestingECG',
    'ExerciseAngina', 'ST_Slope'
])

# Get feature columns (remove target)
X = df_encoded.drop('HeartDisease', axis=1)

print("Total features:", X.shape[1])
print("\nFeature columns:")
for i, col in enumerate(X.columns):
    print(f"  {i+1:2}. {col}")

# ✅ Save — NO RETRAINING NEEDED
with open(r'C:\Users\ABHINAV\Desktop\Heart_PA\models\feature_columns.pkl', 'wb') as f:
    pickle.dump(list(X.columns), f)

print("\n✅ feature_columns.pkl saved!")
```

```python
# Confirm features match model input
model_features = model.input_shape[1]
data_features  = X.shape[1]

if model_features == data_features:
    print(f"✅ MATCH! Model expects {model_features} features")
    print(f"✅ Dataset has {data_features} features")
    print("✅ Ready to use in Streamlit!")
else:
    print(f"❌ MISMATCH!")
    print(f"   Model expects : {model_features}")
    print(f"   Dataset has   : {data_features}")
```

```python
# Load your data + preprocessing
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import numpy as np

# Prepare data
y = df_encoded['HeartDisease']

# Fit scaler on same data
scaler = StandardScaler()
scaler.fit(X)

with open(r'C:\Users\ABHINAV\Desktop\Heart_PA\models\scaler.pkl', 'wb') as f:
    pickle.dump(scaler, f)
print("✅ scaler.pkl saved!")

# Fit KMeans
kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
kmeans.fit(X)

with open(r'C:\Users\ABHINAV\Desktop\Heart_PA\models\kmeans_model.pkl', 'wb') as f:
    pickle.dump(kmeans, f)
print("✅ kmeans_model.pkl saved!")
```
